```python
import matplotlib.pyplot as plt

x = [2, 4, 6, 8, 10]
y = [15, 25, 35, 45, 55]

plt.plot(x, y)
plt.title("Sales Growth")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(["Sales"])
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [95, 88, 76, 69, 54, 40]
scholarship = [6000, 5000, 4000, 2500, 1000, 0]

plt.plot(marks, scholarship)

plt.xlabel("Marks")
plt.ylabel("Scholarship Amount")
plt.title("Scholarship Distribution")
plt.legend(["Scholarship"])

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['P', 'Q', 'R', 'S']
marks = [92, 85, 78, 88]

plt.bar(students, marks)

plt.title("Student Performance")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

salaries = [25000, 27000, 30000, 32000, 35000,
            37000, 40000, 42000, 45000, 47000]

plt.hist(salaries, bins=4)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C', 'JavaScript']
values = [45, 25, 15, 15]

plt.pie(values, labels=labels, autopct='%1.1f%%')

plt.title("Programming Language Popularity")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

plt.subplot(1, 2, 1)
plt.plot(x, [2, 4, 6, 8, 10])
plt.title("Even Numbers")

plt.subplot(1, 2, 2)
plt.plot(x, [1, 3, 5, 7, 9])
plt.title("Odd Numbers")

plt.show()
```
