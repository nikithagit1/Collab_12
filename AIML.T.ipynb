3. Set
Definition

A Set is an unordered collection of unique elements.

Features Unordered Mutable No duplicates allowed Faster searching

```python
numbers = {1, 2, 3, 4}

print(numbers)
```

Important Functions Function Description

add() Add element
remove() Remove element
union() Combine sets
intersection() Common elements
difference() Unique elements
clear() Remove all elements

```python
name1 = {'srija','shiva','teja','priya','Ajith'}
name2 = {'teja','priya','Ajith','kiran','ashok'}
print(name1)
print(name2)
name1.add('task')
print(name1)
name1.remove('srija')
print(name1)
unions = name1.union(name2)
print(unions)
inter = name1.intersection(name2)
print(inter)
dif = name1.difference(name2)
print(dif)
name1.clear()
print(name1)
```

4. Dictionary
Definition

A Dictionary stores data in key-value pairs.

Features Mutable Keys must be unique Fast lookup Stores related information together

```python
student = {
    "name": "John",
    "age": 21,
    "course": "AIML"
}
print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
student.pop('name')
print(student)
student.update({'name':'hello'})
print(student)
student.update({'age':23})
print(student)
```

```python
class Dog:
    species = "Canine"  # Class attribute

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def dog_name(self, dog, cat):
        self.dog = dog
        self.cat = cat
        print("Dog name:", self.dog)
        print("Cat name:", self.cat)


# Taking input from user
name = input("Enter dog object name: ")
age = int(input("Enter age: "))

dog_value = input("Enter dog name: ")
cat_value = input("Enter cat name: ")

# Creating object
d = Dog(name, age)

# Calling function with user input
d.dog_name(dog_value, cat_value)
```
